```python
# Importamos la librería Pandas y la librería Matplotlib

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Cargamos el dataframe y visualizamos las primeras 10 filas

data = pd.read_csv("bank.csv")
data.head(10)
```

## Histograma de frecuencias

* El histograma muestra la frecuencia de cada variable, RECUERDEN que solo mide la frecuencia de una sola variable

```python
plt.subplots(figsize=(10,8))
plt.hist(data["Age"])
plt.xlabel("Edad")
plt.ylabel("Frecuencia")
plt.title("Frecuencia de Edad")

plt.show()
```

```python
plt.subplots(figsize=(10,8))
plt.hist(data["Gender"])
plt.xlabel("Género")
plt.ylabel("Cantidad")
plt.title("Frecuencia de Edad")

plt.show()
```

```python
plt.subplots(figsize=(10,8))
plt.hist(data["Balance"])
plt.xlabel("Saldo en la cuenta")
plt.ylabel("Frecuencia")
plt.title("Frecuencia de saldo en la cuenta")

plt.show()
```

```python
plt.subplots(figsize=(10,8))
plt.hist(data["EstimatedSalary"])
plt.xlabel("Saldo en la cuenta")
plt.ylabel("Frecuencia")
plt.title("Frecuencia de saldo en la cuenta")

plt.show()
```

## Gráfico de dispersión o Scatter Plot

* Este tipo de gráfico muestra la relación entre dos variables

```python
plt.subplots(figsize=(10,8))
plt.scatter(x=data["Age"],y=data["Balance"])
plt.xlabel("Edad")
plt.ylabel("Saldo en la cuenta")
plt.title("Relación entre la Edad y el Saldo en la cuenta")

plt.show()
```

```python
plt.subplots(figsize=(10,8))
plt.scatter(x=data["Age"],y=data["EstimatedSalary"])
plt.xlabel("Edad")
plt.ylabel("Salario Estimado")
plt.title("Relación entre la Edad y el Salario Estimado")

plt.show()
```

```python
plt.subplots(figsize=(10,8))
plt.scatter(x=data["Age"],y=data["NumOfProducts"])
plt.xlabel("Edad")
plt.ylabel("Número de Productos")
plt.title("Relación entre la Edad y el Número de Productos")

plt.show()
```

```python
plt.subplots(figsize=(10,8))
plt.scatter(x=data["Age"],y=data["CreditScore"])
plt.xlabel("Edad")
plt.ylabel("Número de Productos")
plt.title("Relación entre la Edad y el Número de Productos")

plt.show()
```

```python
plt.subplots(figsize=(10,8))
plt.scatter(x=data["CreditScore"],y=data["Balance"])
plt.xlabel("Edad")
plt.ylabel("Número de Productos")
plt.title("Relación entre la Edad y el Número de Productos")

plt.show()
```

## Gráfico de Barras

* Para graficar en barras les enseñaré a utilizar la librería seaborn, la cuál por defecto asignará un color a cada variable
* Este gráfico muestra la frecuencia de una determinada variable en función de otra variable asignada

```python
plt.subplots(figsize=(10,8))
data["Geography"].value_counts().plot(kind="bar", title="Cantidad de clientes por países")
```

```python
plt.subplots(figsize=(10,8))
data["Gender"].value_counts().plot(kind="bar", title="Cantidad de clientes según género")
```

```python
plt.subplots(figsize=(10,8))
plt.bar(x=data["Geography"],height = data["Balance"])
```

```python
plt.subplots(figsize=(10,8))
plt.bar(x=data["Gender"],height = data["Balance"])
```

## Gráfico de Barras usando la librería Seaborn

```python
import seaborn as sns

# Seaborn es una librería creado sobre matplotlib, en la cuál se pueden elaborar gráficos de calidad.
```

```python
plt.figure(figsize=(10,8))
sns.barplot(x=data["Geography"], y = data["Balance"], hue =  data["Gender"])
plt.title("Balance según país y género")
plt.ylabel("Balance")
plt.xlabel("País")
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.barplot(x=data["Geography"], y = data["EstimatedSalary"], hue =  data["Gender"])
plt.title("Balance según país y género")
plt.ylabel("Balance")
plt.xlabel("País")
plt.show()
```
