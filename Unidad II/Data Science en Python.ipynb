### Crear un subconjunto de datos

```python
subset = data["survived"] # Cuando se elije una sola columna se utiliza solo 1 corchete
subset.head()
```

```python
subset2 = data[["survived", "age", "home.dest"]] # Cuando se elijen 2 o más columnas se utilizan 2 corchetes
subset2.head()
```

### Subconjunto de datos con condiciones

```python
cond = data["age"]<18
cond.head()
```

```python
# AND se usa para 

sub_cond = data[(data["survived"]==0) & (data["age"]>18)]
sub_cond.head()
```
